```python
import pandas as p #imported pandas
import numpy as np #imported numpy
from matplotlib import pyplot as plt #imported matplotlib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data = p.read_csv("wdbc.csv",header=None) 
data=data.replace(to_replace="M",value=1) #replaced m as 0 and b as 1
data=data.replace(to_replace="B",value=0)
print(data)
```

```python
y = data.iloc[:,1].values
print(y) #separation of data...output to y variable
```

```python
x=data.iloc[:, 2:32] 
print(x)  #input to x variable
```

```python
from sklearn.model_selection import train_test_split
x_train_i, x_test, y_train_i, y_test = train_test_split(x, y, test_size = 0.10, random_state = 1)
x_train, x_val, y_train, y_val = train_test_split(x_train_i, y_train_i, test_size = 1/9, random_state = 1)
#print(x_train)
print(len(y_train))#splitting the data into three partys namely train, test and validation
#print(X_test)
print(len(y_test))
print(len(y_val))
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)#scaling of data is done
x_test = sc.transform(x_test)
x_val = sc.fit_transform(x_val)
print(x_train)
print(x_test)
print(x_val)
```

```python
def sigmoid_function(Z):
    return 1/(1 + np.exp(- Z))
def predict(X, weights, bias):
    
    m = X.shape[1] #Number of training examples
    y_prediction =  np.zeros((1,m))
    weights = weights.reshape(X.shape[0], 1)
    sigmoid=sigmoid_function(np.dot(weights.T, X)+bias) #agaion control goes to sigmoid function
    for i in range(sigmoid.shape[1]):
        if(sigmoid[0,i]<0.5):
            y_prediction[0,i]=0  #classified into two classses based on the values
        else:
            y_prediction[0,i]=1           
    
    return y_prediction
```

```python
def logistic_regression(xtrain, ytrain, epochs, alpha):
    
    num = xtrain.shape[0] 
    weights = np.zeros((num,1))
    bias = 0   #all essential variables like bias,weights etc.., are all initialized
    fin_costs=[]
    pred=[]
    
    for i in range(epochs):
        m = xtrain.shape[1] #iteration starts here
        Z = np.dot(weights.T, xtrain) + bias;    
        sig = sigmoid_function(Z) #control goes to sigmoid function
        cost= -(1/m) * np.sum(ytrain * np.log(sig) + (1-ytrain) * np.log(1-sig)) #formula of cost function
        d_w = (1/m)* np.dot(xtrain, (sig-ytrain).T)
        d_b = (1/m)* np.sum(sig-ytrain)
        gradients = {"dw" : d_w, "db" : d_b}
        d_w = gradients["dw"]
        d_b = gradients["db"]
        weights = weights - alpha * d_w #updating weights and bias
        bias = bias - alpha * d_b
        if i % 10 == 0:
            fin_costs.append(cost)
            print("cost after %i iteration : %f" % (i, cost)) #cost of every iteration i.e divisible by 20 prints here like 10,20,30
        m = xtrain.shape[1] 
        y_prediction =  np.zeros((1,m))
        weights = weights.reshape(xtrain.shape[0], 1)
        sig_final=sigmoid_function(np.dot(weights.T, xtrain)+bias)
        pred1 = predict(xtrain, weights, bias) #control goes to predict function
        pred.append(100 - np.mean(np.abs(pred1 - ytrain.T)) * 100)
    
   
            
    wb = {"w":weights, "b":bias}
    gradients = {"dw":d_w, "db":d_b}   
    
    
    wght = wb["w"]
    bias = wb["b"]
    
    set_of_values={"weight":weights, "bias":bias, "costs": fin_costs, "predctn": pred}
    
    return set_of_values
```

```python
Logistic_regression_model = logistic_regression(x_train.T, y_train.T, epochs=1000,alpha = 0.01)#first control comes here and logisitc regression function is called

costs=Logistic_regression_model["costs"]
weights=Logistic_regression_model["weight"]
bias=Logistic_regression_model["bias"]
predictions=Logistic_regression_model["predctn"]
```

```python
plt.plot(costs) #plot between costs and number of iterations
plt.title("Cost vs Number of Iterations")
plt.xlabel("Number of Iterations( * 10)")
plt.ylabel("Cost")
```

```python
#print(predictions)
plt.plot(predictions) #plot between prediction and number of iterations
plt.title("Predtictions vs Number of Iterations")
#plt.title("Predictions vs Number of Iterations")
plt.xlabel("Number of Iterations")
plt.ylabel("predictions")
```

```python
y_pred_test = predict(x_test.T, weights, bias)
y_pred_val = predict(x_val.T, weights,bias) #calculation of accuracy
print("\n Validation Data accuracy: {} %".format(100 - np.mean(np.abs(y_pred_val - y_val.T)) * 100)) 
print("\n Test data accuracy: {} %".format(100 - np.mean(np.abs(y_pred_test - y_test.T)) * 100))
```
